```python


import pandas as pd

# Load the main dataset and the survey schema
df = pd.read_csv('results.csv')
schema = pd.read_csv('schema.csv')

# Task 1: Calculate the total number of respondents
total_respondents = len(df)
print(f"Total number of respondents: {total_respondents}")

# Preview the first few rows to understand the data structure
df.head()
```

### Data Cleaning: Identifying Complete Responses

```python

# 1. Get unique question names from the schema
qnames_in_schema = set(schema['qname'].unique())

# 2. Get all column names from our dataset
columns_in_df = set(df.columns)

# 3. Identify common fields (intersection)
# This is crucial as schemas may contain technical fields not present in the results
common_questions = list(qnames_in_schema.intersection(columns_in_df))

# 4. Filter respondents who answered every core question (no NaNs in selected columns)
full_responses = df[common_questions].dropna()

print(f"Number of respondents with complete data: {len(full_responses)}")
```

### Professional Coding Experience Analysis

```python

# Calculate mean, median, and mode for Work Experience
mean_exp = df['WorkExp'].mean()
median_exp = df['WorkExp'].median()
mode_exp = df['WorkExp'].mode()[0] 

# Create a summary series for clean output
stats_table = pd.Series({
    'Mean Experience': mean_exp,
    'Median Experience': median_exp,
    'Mode (Most Common)': mode_exp
})

print("Professional Work Experience Statistics:")
print(stats_table)
```

### Remote Work Count

```python
# Count of Remote Workers
remote_workers = df[df['RemoteWork'] == 'Remote']
total_remote = len(remote_workers)

print(f" Total number of remote workers: {total_remote}")
```

### Python Usage Percentage

```python
# Python Popularity Analysis
# Searching for 'Python' in the technology stack column
python_users = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]

# Calculate global percentage
python_percent = (len(python_users) / len(df)) * 100

print(f"Percentage of respondents using Python: {python_percent:.1f}%")
```

### Online Learning

```python
# Adoption of Online Learning Platforms
online_learners = df[df['LearnCode'].str.contains('Online Courses', na=False)]
print(f"Number of respondents who used Online Courses to learn: {len(online_learners)}")
```

### Top Countries by Python Developer Compensation

```python
# Grouping by country to calculate mean and median yearly compensation for Python users
python_comp_by_country = python_users.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).dropna()

# Sorting by median to identify top-paying markets
python_comp_by_country = python_comp_by_country.sort_values(by='median', ascending=False)

print("Top 10 Countries by Python Developer Median Compensation:")
python_comp_by_country.head(10)
```

### High Earners' Education

```python
# Education Levels of Top 5 Earners
top_5_paid = df.sort_values(by='ConvertedCompYearly', ascending=False).head(5)
result_edu = top_5_paid[['ConvertedCompYearly', 'EdLevel']]

print("Educational Background of the Top 5 Global Earners:")
print(result_edu)
```

### Python Adoption Trends Across Age Categories

```python
# Calculating the percentage of Python developers within each age group
# Note: mean() on a boolean mask calculates the frequency of 'True' values
age_python = df.groupby('Age').apply(lambda x: x['LanguageHaveWorkedWith'].str.contains('Python', na=False).mean() * 100)

print("Python Adoption Rate by Age Group (%):")
print(age_python)
```

### High-Paid Remote Work: Leading Industries

```python

# 1. Define the high-compensation threshold (75th percentile / Q3)
high_comp_threshold = df['ConvertedCompYearly'].quantile(0.75)

# 2. Filter for high-earning remote specialists
high_paid_remote = df[(df['ConvertedCompYearly'] >= high_comp_threshold) & (df['RemoteWork'] == 'Remote')]

# 3. Identify the most frequent industries for these roles
column_to_check = 'Industry' if 'Industry' in df.columns else 'DevType'
top_industries = high_paid_remote[column_to_check].value_counts().head(10)

print(f"Most Popular Industries for High-Compensated Remote Roles:")
print(top_industries)
```
